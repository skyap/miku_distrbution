```python
import skimage as sk
from skimage.io import imread
from skimage.filters import threshold_otsu
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = imread("miku.png")
```

```python
img.shape
```

```python
plt.subplots(figsize=(10,10))
plt.imshow(img)
```

```python
img2 = sk.color.rgba2rgb(img,background=(0.1,0.1,0.1))
img3 = sk.color.rgb2gray(img2)
```

```python
thresh = threshold_otsu(img3)
binary = img3 == 0.1
plt.subplots(figsize=(10,10))
plt.imshow(binary,cmap='Greys')
```

```python
px,py = np.mgrid[0:binary.shape[1]:1,0:binary.shape[0]:1]
x = np.c_[py.ravel(), px.ravel()]
inds_array = np.moveaxis(np.array(x), -1, 0)
y = binary[tuple(inds_array)]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
images = []
max_depth = [5,10,15,20]
for i in range(len(max_depth)):
    clf = RandomForestClassifier(n_estimators=100, max_depth=max_depth[i],random_state = 0,verbose=10, n_jobs = 3)
    clf.fit(x,y)
    preds = clf.predict(x)
    images.append(preds)
```

```python
plt.imshow(preds.reshape(1510,1886).T)
```

```python
fig=plt.figure(figsize=(10, 10))
columns = 2
rows = 2

for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(images[i-1].reshape(1510,1886).T,cmap="binary")
    plt.title(f"max_depth:{max_depth[i-1]}")
fig.suptitle("RandomForestClassifier(n_estimators=100)")
plt.show()
```
